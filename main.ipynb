## Clean + Standardize (Date + Pincode)

```python
# 1) Convert date column to datetime
for df in [total_enrolment, total_demographic, total_biometric]:
    df["date"] = pd.to_datetime(df["date"], format="%d-%m-%Y", errors="coerce")

# 2) Remove rows where date/state/district/pincode is missing
total_enrolment = total_enrolment.dropna(subset=["date", "state", "district", "pincode"])
total_demographic = total_demographic.dropna(subset=["date", "state", "district", "pincode"])
total_biometric = total_biometric.dropna(subset=["date", "state", "district", "pincode"])

# 3) Convert pincode to numeric and then int
for df in [total_enrolment, total_demographic, total_biometric]:
    df["pincode"] = pd.to_numeric(df["pincode"], errors="coerce")

total_enrolment = total_enrolment.dropna(subset=["pincode"])
total_demographic = total_demographic.dropna(subset=["pincode"])
total_biometric = total_biometric.dropna(subset=["pincode"])

total_enrolment["pincode"] = total_enrolment["pincode"].astype(int)
total_demographic["pincode"] = total_demographic["pincode"].astype(int)
total_biometric["pincode"] = total_biometric["pincode"].astype(int)

print(" date & pincode cleaned successfully!")
print("Enrolment shape:", total_enrolment.shape)
print("Demographic shape:", total_demographic.shape)
print("Biometric shape:", total_biometric.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler

# 1) Select key features for risk scoring
risk_features = df_merged[["total_enroll", "update_pressure", "child_ratio"]].copy()

#  2) Normalize to 01 scale
scaler = MinMaxScaler()
scaled = scaler.fit_transform(risk_features)

df_merged["enroll_norm"] = scaled[:, 0]
df_merged["update_pressure_norm"] = scaled[:, 1]
df_merged["child_ratio_norm"] = scaled[:, 2]

#  3) Risk score formula (Weighted)
df_merged["risk_score"] = (
    0.45 * (1 - df_merged["enroll_norm"]) +          # low enrollment => high risk
    0.35 * (df_merged["update_pressure_norm"]) +     # high update pressure => high risk
    0.20 * (1 - df_merged["child_ratio_norm"])       # low child ratio => high risk
)

#  Convert to 0100
df_merged["risk_score_0_100"] = (df_merged["risk_score"] * 100).round(2)

print(" Step 7A done: Risk Score created!")
df_merged[["total_enroll", "update_pressure", "child_ratio", "risk_score_0_100"]].head()
```

### Create Risk Labels (HIGH / MEDIUM / LOW)

```python
#  Quantile-based labeling (data-driven)
q_low = df_merged["risk_score_0_100"].quantile(0.20)
q_high = df_merged["risk_score_0_100"].quantile(0.80)

def label_risk(score):
    if score >= q_high:
        return "HIGH_RISK"
    elif score <= q_low:
        return "LOW_RISK"
    else:
        return "MEDIUM_RISK"

df_merged["risk_label"] = df_merged["risk_score_0_100"].apply(label_risk)

print(" Step 7B done: Risk Labels created!")
print(df_merged["risk_label"].value_counts())
```

### Train Main Model (Risk Classification)

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

#  1) Input Features (X)
features = [
    "total_enroll",
    "total_updates",
    "service_load",
    "update_pressure",
    "child_ratio",
    "bio_ratio"
]

X = df_merged[features]
y = df_merged["risk_label"]

#  2) Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)

print(" Split done!")
print("Train size:", X_train.shape)
print("Test size :", X_test.shape)

#  3) Train Model
clf = RandomForestClassifier(
    n_estimators=200,
    random_state=42,
    class_weight="balanced"
)

clf.fit(X_train, y_train)

#  4) Predictions
y_pred = clf.predict(X_test)

#  5) Evaluation
print("\n Accuracy:", accuracy_score(y_test, y_pred))
print("\n Classification Report:\n", classification_report(y_test, y_pred))

print("\n Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```
